#️⃣ ¿De qué trata este código?
* El programa es una calculadora sencilla para obtener el área y el perímetro de un rectángulo
* Básicamente, te muestra un menú interactivo para que elijas qué quieres calcular
* Lo "chido" es que el código no se cierra solo; se queda abierto en un ciclo para que hagas varios cálculos y tiene un sistema que revisa que no escribas letras o números negativos por error
* #️⃣ Fórmulas que usamosÁrea: Multiplicamos la base por la altura ($A = b \times h$)
* Perímetro: Sumamos los cuatro lados ($P = 2 \times (b + h)$).#️⃣
* ¿Cómo está organizado?validar_datos: Es el filtro que checa que lo que escribas sea un número válido
* calcular_area: Te pide la base y altura para darte el resultado de la superficie
* calcular_perimetro: Te pide los datos para decirte cuánto mide el contorno
* menu: Es el control principal que usa if y elif para decidir qué función llamar según el número que presiones.

Descripción del Problema y del Sistema

* El objetivo de este programa es sistematizar el cálculo de propiedades geométricas de un rectángulo (área y perímetro)
* En el ámbito de la programación inicial, este ejercicio busca resolver la necesidad de procesar datos de entrada del usuario y transformarlos mediante fórmulas matemáticas en información útil
* El sistema actúa como una interfaz de consola interactiva que no solo realiza operaciones aritméticas, sino que también gestiona el flujo de trabajo mediante un ciclo continuo, permitiendo que el usuario realice múltiples consultas sin tener que reiniciar la aplicación cada vez.
Estructura y Modularización (Funciones)

* El código está diseñado bajo el principio de "programación modular"
* Esto significa que cada tarea específica está aislada en su propia función.
* La función validar_datos actúa como un escudo de seguridad, utilizando un bloque try-except o métodos de cadena para verificar que el usuario ingrese números reales y positivos; esto evita que el programa colapse ante entradas erróneas
* Por otro lado, las funciones calcular_area y calcular_perimetro se encargan exclusivamente de la lógica matemática
* Esta separación de responsabilidades hace que el código sea más fácil de leer, mantener y depurar, ya que si hay un error en el cálculo del área, sabemos exactamente en qué sección buscar

Lógica Matemática y Fórmulas Aplicadas

Para la ejecución de los cálculos, el programa solicita dos variables fundamentales: la base (b) y la altura (h)
* El área se define como el espacio bidimensional encerrado dentro de los límites de la figura, y se obtiene mediante la fórmula A=b×h
* El perímetro, que representa la longitud total del contorno del rectángulo, se calcula sumando sus cuatro lados, lo cual se simplifica en la expresión P=2×(b+h)
* El programa utiliza operadores aritméticos estándar de Python (* para multiplicar y + para sumar) y devuelve los resultados con una precisión decimal ajustada para mejorar la legibilidad.

```python
# ############################################################
# ## 🐍 SCRIPT: CALCULADORA GEOMÉTRICA DE RECTÁNGULOS       ##
# ## 👤 NIVEL: PREPARATORIA / PRINCIPIANTE                  ##
# ## ✨ ESTADO: ¡SÚPER CHIDO!                               ##
# ############################################################

# ############################################################
# ## 🛠️ FUNCIÓN: VALIDAR ENTRADA                            ##
# ## Consiste en: No dejar que el usuario rompa el código   ##
# ############################################################
def validar_datos(pregunta):
    while True:
        valor = input(f"✨ {pregunta}")
        # Quitamos puntos para verificar si es un número válido
        if valor.replace('.', '', 1).isdigit():
            num = float(valor)
            if num > 0:
                return num
            else:
                print("❌ ¡No puede ser cero o negativo, carnal!")
        else:
            print("⚠️ ¡Escribe un número de verdad! No letras.")

# ############################################################
# ## 📐 FUNCIÓN: ÁREA DEL RECTÁNGULO                        ##
# ## Fórmula: Área = Base * Altura                          ##
# ############################################################
def calcular_area():
    print("\n   #######################################")
    print("   ##    🧱 MODO: CALCULAR ÁREA 🧱      ##")
    print("   #######################################")
    b = validar_datos("Escribe la BASE: ")
    h = validar_datos("Escribe la ALTURA: ")
    resultado = b * h
    print(f"\n   >> 😎 EL ÁREA ES: {resultado} units² <<")

# ############################################################
# ## 📏 FUNCIÓN: PERÍMETRO DEL RECTÁNGULO                   ##
# ## Fórmula: Perímetro = 2 * (Base + Altura)               ##
# ############################################################
def calcular_perimetro():
    print("\n   #######################################")
    print("   ##   🚧 MODO: CALCULAR PERÍMETRO 🚧   ##")
    print("   #######################################")
    b = validar_datos("Escribe la BASE: ")
    h = validar_datos("Escribe la ALTURA: ")
    resultado = 2 * (b + h)
    print(f"\n   >> 🚀 EL PERÍMETRO ES: {resultado} units <<")

# ############################################################
# ## 🎮 FUNCIÓN: MENÚ PRINCIPAL                             ##
# ## Consiste en: El cerebro que controla todo el flujo    ##
# ############################################################
def menu():
    ejecutando = True

    while ejecutando:
        print("\n" + "#"*40)
        print("##       🔥 GESTOR DE RECTÁNGULOS 🔥    ##")
        print("#"*40)
        print("##  [1] Calcular Área                  ##")
        print("##  [2] Calcular Perímetro             ##")
        print("##  [3] Salir del sistema              ##")
        print("#"*40)

        opcion = input("\n🕹️ Selecciona tu jugada (1-3): ")

        # --- Lógica de decisión ---
        if opcion == "1":
            calcular_area()
        elif opcion == "2":
            calcular_perimetro()
        elif opcion == "3":
            print("\n👋 ¡Cerrando sesión! Qué te vaya chido...")
            ejecutando = False
        else:
            print("\n🚫 Opción inválida. ¡Fíjate bien!")

# ############################################################
# ## 🚀 INICIO DEL PROGRAMA                                 ##
# ############################################################
if __name__ == "__main__":
    menu()
```
